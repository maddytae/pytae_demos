```
import pytae as pt
import numpy as np

fmri = pt.sample_data['fmri']

k = pt.Plotter() #This is how you can initiate a pt plot

(k
.data(fmri) #define data
.plot(x='timepoint', y='signal',by='event',aggfunc='mean',kind='bar',print_data=True)  #define plot usually with x, y and by
.finalize() #finalize the plot
)
k.fig #display or save
```

```
mosaic = """
AB
CC
"""

color={'cue': 'black', 'stim': 'grey', 'parietal': 'blue','frontal': 'green'}


k = pt.Plotter(mosaic)
(k
.data(fmri)
.plot(x='timepoint', y='signal',by='event',aggfunc='mean', on='A',rot=45,kind='bar',
      color=color)
.plot(by='region', on='B')
.plot(by='event', on='C',stacked=True, width=+0.3 ,position=-.3 )
.plot(by='region', on='C', width=+0.3 ,position=1)
.finalize()
)
k.fig
```

```
healthexp= pt.sample_data['healthexp']
mosaic = """
AA
BC
"""
k = pt.Plotter(mosaic)

color={'cue': 'black', 'stim': 'grey', 'parietal': 'blue','frontal': 'green'}
width={'cue':2,'stim':1}
style={'cue':'-','stim':'--'}

(k
.data(healthexp)
.plot(kind='area',x='Year',y='Spending_USD',aggfunc='mean',on='A', by="Country",title='area demo')
.data(fmri)
.plot(x='timepoint', y='signal',by='event',aggfunc='mean',color=color,style=style,width=width,
      on='B',title='combo demo',rot=45,kind='line')
.plot(x='timepoint', y='signal',by='region',kind='bar',stacked=True,on='B',color=color,aggfunc='mean')
 .data(healthexp)
.plot(kind='scatter',x='Life_Expectancy',y='Spending_USD',
      marker='o',s=5,on='C',c='Country',cmap='viridis',title='scatter demo')
.finalize(consolidate_legends=True,ncols=3)
)
k.fig
```

```

tips = pt.sample_data['tips']
healthexp = pt.sample_data['healthexp']



color={'cue': 'blue', 'stim': 'black', 'parietal': 'red',
       'frontal': 'green','Male':'blue','Female':'pink','Dinner':'green',
       'Lunch':'orange','Yes':'grey','No':'brown'}

style={'cue':'-','stim':'--'}
width={'cue':3,'stim':1}
marker = {'Male': 'o', 'Female': 'x'}
size={'Male': 10,'Female': 20}

mosaic = """
ABC
DEF
GHI
JJJ
"""
k = pt.Plotter(mosaic,figsize=(12, 12))

(k
.data(fmri)
.plot(x='timepoint', y='signal',by='event',aggfunc='mean',color=color,style=style,width=width,
      on='A',title='region & event',kind='line')
.plot(x='timepoint', y='signal',by='region',kind='bar',stacked=True,on='A',color=color,aggfunc='mean')
.data(tips)
.plot(x='day',y='total_bill',by='sex',aggfunc='mean',on='B^',kind='bar',color=color,
      title='avg bill by sex')
.plot(x='size',y='tip',by='smoker',on='C',kind='barh',color=color,title='avg tips by smoker status')
.data(tips)
.plot(kind='scatter',x='total_bill',y='tip',on='D',marker='x',s=5,c='day',cmap='viridis',title='scatter plot demo')
.data(tips)
.plot(kind='hexbin',x='total_bill',y='tip',on='E',C='size',reduce_C_function=np.sum,cmap="viridis",gridsize=20,title='hexbin demo')
.data(tips)
.plot(kind='pie',y='tip',by='time',on='F',aggfunc='mean', autopct='%1.1f%%',colors=color,explode=(0,0.05),title='pie demo') #for pie matplotlib says colors and not color
.data(healthexp)
.plot(kind='area',x='Year',y='Spending_USD',aggfunc='mean',on='G^',legend=False, by="Country",title='area plot demo')
.data(tips)

.plot(kind='kde',on='H^',by='day',column='tip',title='kde by days') #if both axes are plotted then legend for right goes to right and vice-versa
.plot(kind='kde',column='tip',on='H', by="sex")
.plot(kind='density',on='I',by='day',column='tip',title='density by days' )
.data(fmri)
.plot(kind='hist',column='signal',by='event',on='J',bins=100)
# .finalize(consolidate_legends=True,ncols=4)
.finalize() #be default legends are always placed in best location as per matplotlib; unless both y axes are used.

)



k.fig

```

```
# Advantages:-

# 1) Easy and consistent chainable syntax: .data() .. .plot() .. .plot() .. .data() .. .plot() .. .finalize(). This is challenging in matplotlib.
# 2) Support for tidy data. This is not available in pandas plot.
# 3) Full integration with matplotlib. This is not available in plotnine. It simply draws an image.
# 4) Since fig is core of the philosophy, it can handle infinite customization in subplots. This is limitation in Seaborn.
# 5) minimal code requirements; all previous .plot() calls are available in subsequent .plot() unless data or the 'kind' of plot is changed. Not available in any package.
# 6) Easy access to underlying data in .plot() using print_data or clip_data. Good for debugging.
# 7) Since it is built on top of matplotlib and pandas plot; Almost all matplotlib or pandas plot **kwargs are supported. This is not possible in Seaborn or plotnine.

# Limitations:-

# 1) Limited statistical features. Check Seaborn or plotnine for that in python or ggplot2 in R
# 2) Since focus is on subplots, faceting is not supported. In fact implicit plots are discouraged by design. Use Seaborn/Plotnine/ggplot2 if implicit faceting is required.

# I look forwards to feedbacks and collaborations.
```
